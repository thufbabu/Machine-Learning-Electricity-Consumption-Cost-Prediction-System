```python
df['water consumption'].fillna(df['water consumption'].mean(), inplace=True)
```

```python
df['utilisation rate'].fillna(df['utilisation rate'].mean(), inplace=True)
```

```python
df = df.drop(['Maintenance_Status', 'Energy_Source','Backup_Power'], axis=1)
```

```python
df.isnull().sum()/len(df)*100
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Meter_Type']=le.fit_transform(df[['Meter_Type']])
```

```python
from sklearn.preprocessing import LabelEncoder
le1=LabelEncoder()
df['Peak_Hour_Usage']=le1.fit_transform(df[['Peak_Hour_Usage']])
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe1=OneHotEncoder(sparse_output=False)
ohe1.fit(df[['Appliance_Type']])
df1=ohe1.transform(df[['Appliance_Type']])

df1=pd.DataFrame(df1,columns=ohe1.get_feature_names_out())
```

```python
df=pd.concat([df,df1],axis=1)
```

```python
df['Usage_Pattern']=df['Usage_Pattern'].replace({'High':0,'Medium':1,'Low':2})
```

```python
df['structure type'] = df['structure type'].replace({'Mixed-use': 0, 'Residential': 1, 'Commercial': 2, 'Industrial': 3})
```

```python
df['Weather_Condition']=df['Weather_Condition'].replace({'Rainy':0,'Winter':1,'Summer':2})
```

```python
df.head()
```

```python
df=df.drop('Appliance_Type',axis=1)
```

```python
x=df.drop('electricity cost',axis=1)
y=df['electricity cost']
```

```python
# x= pd.get_dummies(x, drop_first=True)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

MODEL BUILDING

```python

from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
```

```python
R2= r2_score(y_test,y_pred)
R2
```

```python
y_pred = lr.predict(x_train)
y_pred
```

```python
R2= r2_score(y_train,y_pred)
R2
```

```python
# Train performance
y_pred_train = lr.predict(x_train)
print("Train MAE:", mean_absolute_error(y_train, y_pred_train))

# Test performance
y_pred_test = lr.predict(x_test)
print("Test MAE :", mean_absolute_error(y_test, y_pred_test))
```

```python
# Train performance
y_pred_train = lr.predict(x_train)
print("Train MAE:", mean_squared_error(y_train, y_pred_train))

# Test performance
y_pred_test = lr.predict(x_test)
print("Test MAE :", mean_squared_error(y_test, y_pred_test))
```

KNEIGHBORSREGRESSOR

```python
from sklearn.neighbors import KNeighborsRegressor
knr=KNeighborsRegressor()
knr.fit(x_train, y_train)
```

```python
y_pred = knr.predict(x_test)

print("Train Accuracy:", knr.score(x_train, y_train))
print("Test Accuracy:", r2_score(y_test, y_pred))
```

RANDOMFORESTREGRESSOR

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
rfr.fit(x_train, y_train)
```

```python
y_pred = rfr.predict(x_test)
print("Train R2:", r2_score(y_train, rfr.predict(x_train)))
print("Test R2 :", r2_score(y_test, y_pred))

print("MAE :", mean_absolute_error(y_test, y_pred))
print("RMSE:", mean_squared_error(y_test, y_pred))
```
