```python
import numpy as np
import pandas as pd
import pickle
```

```python
lr = pickle.load(open("lr_model.sav", "rb"))
scaler = pickle.load(open("scaler.sav", "rb"))
ohe1 = pickle.load(open("Appliance_Type.sav", "rb"))
le = pickle.load(open("Meter_Type.sav", "rb"))
le1 = pickle.load(open("Peak_Hour_Usage.sav", "rb"))
```

```python
site_area = float(input("Enter site area: "))
water_consumption = float(input("Enter water consumption: "))
recycling_rate = float(input("Enter recycling rate: "))
utilisation_rate = float(input("Enter utilisation rate: "))
air_quality_index = float(input("Enter air quality index: "))
issue_resolution_time = float(input("Enter issue resolution time: "))
resident_count = int(input("Enter resident count: "))
structure_type = input("Enter structure type (Mixed-use / Residential / Commercial / Industrial): ")
appliance_type = input("Enter appliance type (AC / Fan / Heater / etc): ")
meter_type = input("Enter meter type (Smart / Analog): ")
usage_pattern = input("Enter usage pattern (High / Medium / Low): ")
peak_hour_usage = input("Enter peak hour usage (Yes / No): ")
weather_condition = input("Enter weather condition (Rainy / Winter / Summer): ")
```

```python
#  2127	Residential	1489.0	15	65.0	117	52	124	2656	Refrigerator	Analog	High	No	Rainy
```

```python
data= pd.DataFrame([{
    'site area': site_area,
    'structure type': structure_type,
    'water consumption': water_consumption,
    'recycling rate': recycling_rate,
    'utilisation rate': utilisation_rate,
    'air qality index': air_quality_index,
    'issue reolution time': issue_resolution_time,
    'resident count': resident_count,
    'Appliance_Type': appliance_type,
    'Meter_Type': meter_type,
    'Usage_Pattern': usage_pattern,
    'Peak_Hour_Usage': peak_hour_usage,
    'Weather_Condition': weather_condition
}])
```

```python
# Usage Pattern
data['Usage_Pattern'] = data['Usage_Pattern'].replace({
    'High': 0,
    'Medium': 1,
    'Low': 2
})

# Structure Type
data['structure type'] = data['structure type'].replace({
    'Mixed-use': 0,
    'Residential': 1,
    'Commercial': 2,
    'Industrial': 3
})

# Weather Condition
data['Weather_Condition'] = data['Weather_Condition'].replace({
    'Rainy': 0,
    'Winter': 1,
    'Summer': 2
})
```

```python

# Label Encoding
data['Meter_Type'] = le.transform(data[['Meter_Type']])
data['Peak_Hour_Usage'] = le1.transform(data[['Peak_Hour_Usage']])

# One Hot Encoding Appliance_Type
appliance_encoded = ohe1.transform(data[['Appliance_Type']])
appliance_df = pd.DataFrame(
    appliance_encoded,
    
    columns=ohe1.get_feature_names_out()
)

data = pd.concat(
    [data.drop('Appliance_Type', axis=1), appliance_df],
    axis=1
)

# Scaling
data = scaler.transform(data)

# Prediction
prediction = lr.predict(data)
print("Predicted Electricity Cost:", prediction[0])
```
